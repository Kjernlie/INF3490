```python
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns

%matplotlib inline

def function(x):
    return -x**4 + 2*x**3 + 2*x**2 - x

def f_derivative(x):
    return -4*x**3 + 6*x**2 + 4*x -1

N=20
x = np.linspace(-2,3,20)
f = np.zeros(N)
f_der = np.zeros(N)

for i in range(len(x)):
    f[i] = function(x[i])
    f_der[i] = f_derivative(x[i])


plt.plot(x,f,label='f')
plt.plot(x,f_der,label='f_der')
plt.legend()
```

```python
# Gradient ascent

x_old = 0 # The value does not matter as long as abs(x_new - x_old) > precision
x_new = -1 # Where the algorithm starts.
gamma = 0.1 # step size
precision = 0.00001

#f_derivative(-1)
while abs(x_new - x_old) > precision:
    x_old = x_new
    x_new = x_old + gamma * f_derivative(x_old)
    x_new

print("Local maximum occurs at", x_new)
```

```python
# exhaustive search

import numpy as np
h=0.5
N=(2+3)/h
x = np.linspace(-2,3,N)
a_old=0

for i in range(len(x)):
    a_new = function(x[i])
    if a_new > a_old:
        x_max = x[i]

print x_max
        
    
```
